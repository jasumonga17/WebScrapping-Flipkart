```python
#Necessary Libraries
from bs4 import BeautifulSoup as soup
import requests
import pandas as pd
import csv
```

```python
# Data on each page
# After Adding the URL. In the end add ...&page'+str(i)
# This will give data over each page.


page_num=input("Enter Page No:")
for i in range(1,int(page_num)+1):
    #URL
    url='https://www.flipkart.com/search?p%5B%5D=facets.brand%255B%255D%3DSamsung&sid=tyy%2F4io&sort=recency_desc&wid=1.productCard.PMU_V2_1&page='+str(i)
    req=requests.get(url)
    content=soup(req.content,'html.parser')
    print(content)
        
```

```python
# Creating Empty List and giving the class for each category
phn_name=[]
phn_price=[]
phn_rating=[]
name=content.find_all('div',{'class':'_4rR01T'})
price=content.find_all('div',{'class':'_30jeq3 _1_WHN1'})
rating=content.find_all('div',{'class':'_3LWZlK'})
print('Phone on Page:'+str(i))
print(len(name))
```

```python
# Storing the Records in an empty list
for i in name:
    phn_name.append(i.text)
for i in price:
    phn_price.append(i.text)
for i in rating:
    phn_rating.append(i.text)
```

```python
# DataFrame

# Based on the page data will be shown

data=pd.DataFrame({'Phone Name':phn_name,'Price':phn_price,'Rating':phn_rating})
data.head()
```
